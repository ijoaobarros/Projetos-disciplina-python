```python
#Importações de bibliotecas e criação de funções

import pandas as pd
import plotly.express as px

def Maxima_Potencia(dados):
    potencias = dados['Potência']
    tam_potencias = len(potencias)
    i = tam_potencias-2     #Para iniciar no penúltimo item do array
    max_pot = potencias.iloc[tam_potencias-1]

    while True:
        if max_pot<potencias.iloc[i]:
            max_pot = potencias.iloc[i]
            i-=1
        elif max_pot<potencias.iloc[i-1]:
            max_pot = potencias.iloc[i-1]
            i-=2
        elif max_pot<potencias.iloc[i-2]:
            max_pot = potencias.iloc[i-2]
            i-=3
        elif max_pot<potencias.iloc[i-3]:
            max_pot = potencias.iloc[i-3]
            i-=4
        elif max_pot<potencias.iloc[i-4]:
            max_pot = potencias.iloc[i-4]
            i-=5
        elif max_pot<potencias.iloc[i-5]:
            max_pot = potencias.iloc[i-5]
            i-=6
        elif max_pot<potencias.iloc[i-6]:
            max_pot = potencias.iloc[i-6]
            i-=7
        else:
            break
    return max_pot

str_as_float = lambda x: float(x)       #Função para converter string em float

def arquivo_em_lista(nome_arquivo: str):
    arquivo = open(nome_arquivo, 'r')
    lista = list(map(str_as_float, arquivo.readlines()[0].split(",")))
    arquivo.close()
    return lista
```

```python
nome_do_arquivo_V = ['V_RSM100P_S361_T45.txt','V_RSM100P_S589_T46.txt','V_RSM100P_S720_T41.txt','V_RSM100P_S901_T46.txt']
nome_do_arquivo_I = ['I_RSM100P_S361_T45.txt','I_RSM100P_S589_T46.txt','I_RSM100P_S720_T41.txt','I_RSM100P_S901_T46.txt']

V_Curva, I_Curva = [], []

#Converter todos os arquivos em listas
for i in range(len(nome_do_arquivo_V)):
    V_Curva.append(arquivo_em_lista(nome_do_arquivo_V[i]))
    I_Curva.append(arquivo_em_lista(nome_do_arquivo_I[i]))
```

```python
#Criar lista de dataframes
df = []
for i in range(len(nome_do_arquivo_V)):
    df_temp = pd.DataFrame(V_Curva[i], columns=['Tensão'])
    df_temp['Corrente'] = I_Curva[i]
    df_temp['Potência'] = df_temp['Tensão'] * df_temp['Corrente']
    df.append('')
    df[i] = df_temp
    print(f'Para a medição {nome_do_arquivo_V[i][10:18]}:')
    display(df[i])
```

```python
for i in range(len(nome_do_arquivo_V)):
    titulo = nome_do_arquivo_V[i][10:18]
    grafico = px.line(df[i], x='Tensão', y='Potência', title=titulo)

    #Encontrar ponto de máxima potência
    max_pot = Maxima_Potencia(df[i])
    temp = df[i].set_index('Potência')
    max_tensao_associada = temp.loc[max_pot, 'Tensão']  #Encontra a tensão associada a máxima potência

    #Marcar no gráfico o ponto de máxima potência -> (max_tensao_associada, max_pot)
    grafico.add_scatter(x=[max_tensao_associada], y=[max_pot], mode='markers', marker=dict(color='red', size=10), name='Máxima Potência')
    grafico.show()
```
